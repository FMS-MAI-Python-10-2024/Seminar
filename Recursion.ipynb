Игра завершается в тот момент, когда количество камней в куче становится числом, оканчивающимся на ноль. Победителем считается игрок, сделавший последний ход, т. е. первым получивший кучу, количество камней в которой оканчивается на ноль. К примеру, игра заканчивается, когда в куче стало 10, 200, 6800 камней.

Будем говорить, что игрок имеет выигрышную стратегию, если он может выиграть при любых ходах противника.

```python
def solve_stones_1(move, stones):
    if stones % 10 == 0:
        if move == 2:
            return True
        else:
            return False
        
    if move > 2:
        return False
        
    if move % 2 == 0:
        return solve_stones_1(move + 1, stones + 1) and solve_stones_1(move + 1, stones + 3) and solve_stones_1(move + 1, stones + 11)
    else:
        return solve_stones_1(move + 1, stones + 1) or solve_stones_1(move + 1, stones + 3) or solve_stones_1(move + 1, stones + 11)

# 16
for i in range(11, 99):
    if i % 10 != 0:
        if solve_stones_1(0, i):
            print(i)
            break
```

+ Прибавить 3.

```python
nums = set()
def solve_23(start, nums):
    if start % 2 != 0 and start < 100:
        nums.add(start)
    if start > 100:
        return
    
    solve_23(start + 3, nums)
    solve_23(start * 3, nums)


solve_23(10, nums)
print(f"Количество искомых программ: {len(nums)}") # 36
```

### Задание 16.

Алгоритм вычисления значения функции F(n), n — натуральное число, задан следующими соотношениями:
+ F(n)=1 при n=1;
+ F(n)=(n−1)×F(n−1), если n>1.

Чему равно значение выражения (F(2024)+2×F(2023))/F(2022)?

```python
answers = dict()

def f(n):
    if n in answers:
        return answers[n]
    if n == 1:
        answers[n] = 1
        return 1
    answers[n] = (n - 1) * f(n - 1)
    return answers[n]


print((f(2024) + (2 * f(2023))) / f(2022)) # 4094550
```

+ F(n)=n⋅2 при n≤10
+ F(n)=F(n−3)−F(n−9)⋅2, если n чётно и n>10
+ F(n)=F(n−2)⋅2−F(n−7), если n нечётно и n>10

Определите сумму цифр результата для F(3063).

### Задание 16.

Алгоритмы вычисления значения функций F(n) и G(n) заданы следующими соотношениями:
+ F(n)=1, если n≥3210,
+ G(n)=n, если n<10,
+ F(n)=F(n+3)+7, если n<3210,
+ G(n)=G(n−3)+5, если n≥10.

Чему равно значение выражения F(15)−G(3000)?

```python
dict_f = dict()
dict_g = dict()

def f(n):
    if n in dict_f:
        return dict_f[n]
    if n >= 3210:
        dict_f[n] = 1
        return dict_f[n]
    if n < 3210:
        dict_f[n] = f(n + 3) + 7
        return dict_f[n]


def g(n):
    if n in dict_g:
        return dict_g[n]
    if n < 10:
        dict_g[n] = n
        return dict_g[n]
    if n >= 10:
        dict_g[n] = g(n - 3) + 5
        return dict_g[n]
    
    
print(f(15) - g(3000)) # 2462
```

Алгоритмы вычисления значения функций F(n) и G(n), где n — натуральное число, задан следующими соотношениями:
+ G(n)=F(n)=1 при n<3;
+ F(n)=G(n)+F(n−1) при n>2 и n чётно;
+ F(n)=F(n−2)−2⋅G(n+1) при n>2 и n нечётно;
+ G(n)=F(n−3)+F(n−2) при n>2 и n чётно;
+ G(n)=F(n+1)−G(n−1) при n>2 и n нечётно.

Чему равно значение функции G(120)? В ответе запишите только целое число.

```python
dict_f.clear()
dict_g.clear()

def F(n):
    if n in dict_f:
        return dict_f[n]
    if n < 3:
        dict_f[n] = 1
        return dict_f[n]
    if n > 2 and n % 2 == 0:
        dict_f[n] = G(n) + F(n - 1)
        return dict_f[n]
    if n > 2 and n % 2 != 0:
        dict_f[n] = F(n - 2) - 2 * G(n + 1)
        return dict_f[n]
    

def G(n):
    if n in dict_g:
        return dict_g[n]
    if n < 3:
        dict_g[n] = 1
        return dict_g[n]
    if n > 2 and n % 2 == 0:
        dict_g[n] = F(n - 3) + F(n - 2)
        return dict_g[n]
    if n > 2 and n % 2 != 0:
        dict_g[n] = F(n + 1) - G(n - 1)
        return dict_g[n]
    
    
print(G(120)) # 118
```
